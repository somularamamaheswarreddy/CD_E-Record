```python

def generate_quadruples(expr):
    temp = 1
    quads = []
    stack = []
    output = []

    # Step 1: Convert infix → postfix (very simple)
    precedence = {'+':1, '*':2}
    operators = []

    tokens = expr.split()

    for token in tokens:
        if token.isalnum():
            output.append(token)
        elif token in precedence:
            while (operators and
                   precedence.get(operators[-1], 0) >= precedence[token]):
                output.append(operators.pop())
            operators.append(token)

    while operators:
        output.append(operators.pop())

    # Step 2: Generate quadruples from postfix
    for token in output:
        if token.isalnum():
            stack.append(token)
        else:
            op2 = stack.pop()
            op1 = stack.pop()
            temp_var = f"T{temp}"
            temp += 1
            quads.append((token, op1, op2, temp_var))
            stack.append(temp_var)

    return quads


print(generate_quadruples("burger + fries * 2"))
```
